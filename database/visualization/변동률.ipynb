```python
import pandas as pd
df = pd.read_csv('./시계열매출_매출건수_점포.csv',encoding='utf-8')
```

```python
#매출액 음수값 제외
df[df['점포당평균매출액'] <= 0]
```

```python
#변동률 계산을 위한 연도별 분류
df2021 = df[(df['연도']=='2021Y 1Q') | (df['연도']=='2021Y 2Q')]
df2020 = df[(df['연도']=='2020Y 1Q') | (df['연도']=='2020Y 2Q')]
df2021 = df2021.groupby(['구명','행정동명','서비스업종명'],as_index=False)['점포당평균매출액'].sum()
df2021['연도'] = 2021
df2020 = df2020.groupby(['구명','행정동명','서비스업종명'],as_index=False)['점포당평균매출액'].sum()
df2020['연도'] = 2020
df2021 = df2021[['연도','구명','행정동명','서비스업종명','점포당평균매출액']] 
df2020 = df2020[['연도','구명','행정동명','서비스업종명','점포당평균매출액']]
```

```python
df2021
```

```python
df2020
```

```python
df2021 = pd.merge(df2021,df2020,on=['구명','행정동명','서비스업종명'])
df2021
```

```python
#변동률 계산
df2021['변동률'] = (df2021['점포당평균매출액_x'] - df2021['점포당평균매출액_y'] )/ df2021['점포당평균매출액_y'].abs() * 100
df2021
```

```python
df2021.columns = ['연도','구명','행정동명','서비스업종명','점포당평균매출액','이전연도','이전점포당평균매출액','변동률']
df2021 = df2021.astype({'변동률':'int'})
```

```python
df2021.to_csv('./변동률.csv',encoding='utf-8-sig',index=False)
```
