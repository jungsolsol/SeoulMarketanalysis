```python
import pandas as pd
import matplotlib.pyplot as plt
population = pd.read_csv('./시계열일일생활인구.csv',encoding='utf-8')
```

```python
population.columns
```

```python
#자치구별 groupby
gu_pop = population.groupby(['연도','구명'],as_index=False)['일일인구','10대인구', '20대인구', '30대인구', '40대인구', '50대인구',
       '60대이상인구', '월요일인구', '화요일인구', '수요일인구', '목요일인구', '금요일인구', '토요일인구',
       '일요일인구', '남성인구', '여성인구'].sum()
```

```python
gu_pop.to_csv("./구단위인구.csv",encoding='utf-8-sig',index=False)
```

```python
#생활인구 데이터 분석을 위한 시각화 
import seaborn as sns
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(32, 10))ㄴㄴㄴㄴ
sns.barplot(data=gu_pop, x='연도', y='일일인구', hue='구명')
plt.legend(fontsize=14)
plt.show()
```
