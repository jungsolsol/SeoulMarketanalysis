```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
df = pd.read_csv('/Users/krc/Downloads/gitfile/변동률.csv',encoding='utf-8')
```

```python
df.head
```

```python

df_local = df.groupby(['구명','행정동명'],as_index=False)['점포당평균매출액','이전점포당평균매출액',].sum()
df_local['변동률'] = (df_local['점포당평균매출액'] - df_local['이전점포당평균매출액']) / df_local['이전점포당평균매출액'] * 100
df_local
```

```python
df_service = df.groupby(['서비스업종명'],as_index=False)['점포당평균매출액','이전점포당평균매출액',].sum()
df_service['변동률'] = (df_service['점포당평균매출액'] - df_service['이전점포당평균매출액']) / df_service['이전점포당평균매출액'] * 100
df_service
```

```python
df_local = df_local.astype({'변동률':int})
df_service = df_service.astype({'변동률':int})
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df_local['행정동명'],df_local['변동률'],s=20)
```

```python
#정규화
from scipy import stats
plt.figure(figsize=(20,8))
df_local['변동률'] = stats.zscore(df_local['변동률'])
df_local = df_local[df_local['변동률'].between(-2,2)]
plt.scatter(df_local['행정동명'],df_local['변동률'],s=10)
df_local.to_csv("./정규지역변동률.csv",encoding='utf-8-sig',index=False)
```
