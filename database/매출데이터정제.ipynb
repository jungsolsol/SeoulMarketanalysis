```python
import pandas as pd
df = pd.read_csv("/Users/krc/hyeongJin/데이터원본파일/원본통합.csv",encoding='utf-8')
location = pd.read_csv('/Users/krc/hyeongJin/project-template/database/HyeongJin/상권행정동추가.csv',encoding='utf-8')
```

```python
df.columns
```

```python
df_sales = df[['기준_년_코드', '기준_분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명',
       '서비스_업종_코드', '서비스_업종_코드_명', '분기당_매출_금액', '분기당_매출_건수','점포수']]
```

```python
df_sales = df_sales[df_sales['상권_구분_코드_명'] != '관광특구']
```

```python
df_sales = df_sales[df_sales['기준_년_코드'] != 2017]
```

```python
df_sales
```

```python
location
```

```python
sales_mer = pd.merge(df_sales,location,on='상권_코드').drop_duplicates().reset_index()
```

```python
sales_mer.columns
```

```python
sales_mer = sales_mer[['기준_년_코드', '기준_분기_코드','시군구명','법정동','상권_코드_명_x','서비스_업종_코드_명', '분기당_매출_금액', '분기당_매출_건수', '점포수']]
```

```python
sales_mer.columns = ['기준년코드','기준분기코드','구명','행정동명','상권코드명','서비스업종명','분기당매출금액','분기당매출건수','점포수']
```

```python
sales_mer
```

```python
sales_mer = sales_mer[sales_mer['점포수']!=0]
sales_mer['년도'] = (sales_mer['기준년코드'].astype(str)) + 'Y' + (sales_mer['기준분기코드'].astype(str)) + 'Q'
```

```python
sales_mer = sales_mer[['년도','구명','행정동명','상권코드명','서비스업종명','분기당매출금액','분기당매출건수','점포수']]
sales_mer
```

```python
sales = sales_mer.groupby(['년도','구명','행정동명','서비스업종명'],as_index=False)['분기당매출금액','분기당매출건수','점포수'].sum()
sales
```

```python
sales = sales[~sales['년도'].str.contains('2018')]
```

```python
sales = sales.astype({'분기당매출금액':int})
```

```python
sales['점포당평균매출액'] = (sales['분기당매출금액']/sales['점포수']).astype(int)
```

```python
sales
```

```python
sales.to_csv('./1201.csv',encoding='utf-8-sig',index=False)
```
